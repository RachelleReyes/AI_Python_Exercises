```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Generación de los 50 puntos en el rango [-10,10]
x = np.linspace(-10, 10, num=50)
print('50 Puntos en un rango de x [-10,10] ')
print(x)

m = [0.5,1,1.5,2,0.5,1,1.5,2] #Valores de media
s = [1,2,3,4,4,3,2,1] #Valores de desviación

#Valores para acomodar graficas de 2x4
renglon = 2 
columna = 4
posx = 0
posy = 0

# Funcion que muestra varias gráficas en una ventana
fig, axs = plt.subplots(renglon,columna,figsize=(20,10))

#Ciclo que evalúa la función y que lo grafica
for i in range(len(m)):
    y = []
    for j in range(len(x)):
        # Evaluación de la función
        y+= [1/np.sqrt(2*np.pi)*np.exp(-1/2*((x[j]-m[i])/s[i])**2),]
    
    # Creación de las gráficas
    titulo = '\nMedia: '+ str(m[i]) + ' ' + 'Desviación: '+ str(s[i])
    print(titulo)
    print(y)
    axs[posx, posy].scatter(x,y)  #Función que crea la gráfica
    
    axs[posx, posy].set(xlabel ='x',ylabel = 'f(x)',title = titulo)
    axs[posx, posy].grid()
    
    posy+= 1
    if posy == columna:        #Comparación para la posición de la gráfica
        posx+=1
        posy = 0
    del y
        
fig.savefig("ejercicio1.png")
plt.show()
```
